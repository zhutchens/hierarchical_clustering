```python
# csv file header: text, book, chapter. im only interested in the first book for now, csv file includes every book

# end goal is to split the book into a set of chapters, and each chapter has a set of sentences 

# read csv file using pandas and construct DataFrame 
dataframe = pd.read_csv("../data/harry_potter_books.csv")

# list of harry potter books 
book_list = dataframe["book"].unique().tolist()

# have user enter book to generate hierarchy for
user_book = ''
while user_book not in book_list:
    user_book = input(f'Enter a book from this list exactly as it is: {book_list}')
    if user_book not in book_list:
        user_book = input(f'Book not in list, try again: {book_list}')

book_df = dataframe.query(f'book == "{user_book}"')

# retrieve chapter title list 
chapter_title_list = chapter_titles[book_list.index(user_book)]

chapter_list = []
sentences_list = []
chapter_text = ''
for chapter in book_df["chapter"].unique():
    chapter_text = ''
    for text in book_df.query(f"chapter == '{chapter}'")["text"].values:
        chapter_text += text + ' '
    chapter_list.append(chapter_text)
    sentences_list.append(book_df.query(f"chapter == '{chapter}'")["text"].values.tolist())
        
```

```python
# display word relations on given sentence

# retrieve list 
all_book_sentences = []
for sentence in book_df['text'].values:
    all_book_sentences.append(sentence)

user_sentence = -1
while user_sentence not in range(len(all_book_sentences)):
    try:
        user_sentence = int(input(f'Enter a number in the range 0 - {len(all_book_sentences)}'))
    except ValueError:
        user_sentence = int(input(f'Invalid selection, enter a number in the range 0 - {len(all_book_sentences)}'))

import spacy
nlp = spacy.load('en_core_web_lg')
doc = nlp(all_book_sentences[user_sentence])
spacy.displacy.render(doc, style = 'dep', jupyter = True)
```

```python
ontology_hierarchy, relation, titles = construct_topic_modeling_concept_hierarchy(
    chapters = chapter_list,
    titles = chapter_title_list,
    sentences_per_chapter = sentences_list,
    # how many number of topics should there be relative to number of chapters?
    num_topics = int(len(chapter_title_list) // 3),
    chosen_cluster = 1
)

draw_hierarchy_tree_from_ontology (
    ontological_hierarchy = ontology_hierarchy,
    relations_to_verbs = relation,
    topic_modelling_chapters = chapter_title_list
)
```
